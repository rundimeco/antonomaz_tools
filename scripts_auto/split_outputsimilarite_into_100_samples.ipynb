# echantillon100_output similarite
* Ce script sert à découper le fichier "sortie_similarite.txt" (output du Search_pdf...ipynb) en échantillons de 100 sorties, pour myriadiser la relecture humaine.

```python
import re
import glob
```

```python
with open("sortie_similarite.txt", "r", encoding="utf8") as f:
    texte=f.read()
```

```python
print(texte)
```

```python
liste_entites=re.split("------------------------------", texte)
```

```python
total=len(liste_entites)
print(total)
for i in range(0, total,100):
    echantillon=liste_entites[i:i+100]
    for e in echantillon:
        with open("echantillon_%s.txt" %i, "a", encoding="utf8") as f:
            f.write(e + "\n"+"------------------------------"+"\n")
        
```
